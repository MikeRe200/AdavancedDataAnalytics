```python
filepath='datasets/12dancingprincesses.txt'
```

```python
with open(filepath, 'r') as f:
    for line in f:
        print(line)
```

```python
List=[]

with open(filepath, 'r') as f:
    for line in f:
        cline = line.strip() #get rid of newline character
        #print(cline)
        if cline == '': pass  #this will skip over lines that only had a newline and are now blank
        else:
            tknls = word_tokenize(cline)   #the comma is now its own token when the sentence is split
            
            for token in tknls:
                List.append(token)
```

```python
len(List)
```

```python
#Remove the punctuation
#remove the puntuation tokens from the list
for word in List:
    if word in punctuation:
            List.remove(word)
```

```python
len(List)
```

```python
from nltk.corpus import stopwords
rm_count = 0
new_words = []  #list to hold new words
eng_stopwords = stopwords.words("english")
for word in List:
    if word not in eng_stopwords:
        new_words.append(word)
    else: rm_count += 1
```
